Почему выбрали эти модели:

* Linear Regression – это базовая модель, которая проверяет наличие линейной зависимости между признаками. Она помогает быстро и просто оценить влияние факторов на целевой признак.

* Random Forest Regressor – данная модель была выбрана из-за способности находить нелинейные зависимости в данных. А так же это ансамблевая модель, которая строит множество деревьев решений и объединяет их для улучшения точности.

### Подготовка данных для обучения

```python
# Замена значений признаков на числовые, то есть вместо условной строки «8 GB», будет числовое значение 8
df_laptopPrice['ram_gb'] = df_laptopPrice['ram_gb'].astype(str).str.replace('GB', '').astype(int)
df_laptopPrice['ssd'] = df_laptopPrice['ssd'].astype(str).str.replace('GB', '').astype(int)
df_laptopPrice['hdd'] = df_laptopPrice['hdd'].astype(str).str.replace('GB', '').astype(int)
df_laptopPrice['graphic_card_gb'] = df_laptopPrice['graphic_card_gb'].str.replace('GB', '').astype(int)
df_laptopPrice['os_bit'] = df_laptopPrice['os_bit'].str.replace('-bit', '').astype(int)
df_laptopPrice['processor_gnrtn'] = df_laptopPrice['processor_gnrtn'].str.replace('th', '')
df_laptopPrice['rating'] = df_laptopPrice['rating'].str.replace(' stars', '', regex=False)
df_laptopPrice['rating'] = df_laptopPrice['rating'].str.replace(' star', '', regex=False)
df_laptopPrice['processor_gnrtn'] = pd.to_numeric(df_laptopPrice['processor_gnrtn'], errors='coerce')
median_value = df_laptopPrice['processor_gnrtn'].median()

df_laptopPrice['processor_gnrtn'] = df_laptopPrice['processor_gnrtn'].fillna(median_value)
df_laptopPrice['processor_gnrtn'] = df_laptopPrice['processor_gnrtn'].astype(int)
```

Память видеокарты (graphic_card_gb):

Умеренная зависимость с ценой (0.46) и объемом SSD (0.33), что ожидаемо — более мощные видеокарты ставятся в ноутбуки премиум-сегмента.

```python
# Замена оставшихся текстовых полей на числовые значения
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
for column in df_laptopPrice.columns:
    if df_laptopPrice[column].dtype == 'object':
        le = LabelEncoder()
        df_laptopPrice[column] = le.fit_transform(df_laptopPrice[column])
        label_encoders[column] = le
```

### Обучение моделей

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df_laptopPrice.fillna(df_laptopPrice.median(), inplace=True)
```

```python
X = df_laptopPrice.drop(columns=['Price'])
y = df_laptopPrice['Price']
```

```python
# Стандартиация
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### LinearRegression

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

y_pred_lr = lr_model.predict(X_test)

mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print(f"mean-squared-error: {mse_lr}")
print(f"r2score: {r2_lr}")
```

```python
plt.scatter(y_test,y_pred_lr)
```

### RandomForestRegressor

```python
rfr = RandomForestRegressor(random_state=42)
rfr.fit(X_train, y_train)

y_pred_rfr = rfr.predict(X_test)

mse_rfr = mean_squared_error(y_test, y_pred_rfr)
r2_rfr = r2_score(y_test, y_pred_rfr)
import math
print(f"mean-squared-error: {mse_rfr}")
print(f"r2score: {r2_rfr}")
```

```python
plt.scatter(y_test,y_pred_rfr)
```

### DecisionTreeRegressor

```python
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)

y_pred_dt = dt_model.predict(X_test)

mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

print(f"mean-squared-error: {mse_dt}")
print(f"r2score: {r2_dt}")
```

```python
plt.scatter(y_test,y_pred_dt)
```
